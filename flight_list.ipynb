```python
import pandas as pd
import tabula.io as tb

```

```python
link = 'https://ia803407.us.archive.org/33/items/epstein-flight-logs-unredacted/EPSTEIN%20FLIGHT%20LOGS%20UNREDACTED.pdf'

df = tb.read_pdf(link, stream = True, pages='all')
```

```python
####JOINS ENTIRE DATAFRAME TOGETHER#######

#df2 = df[0] #opens dataframe
length = len(df)

df2 = pd.DataFrame()

for i in range(length+1):

    if i < length - 2: 
        df2 = pd.concat([df2, df[i+1]],axis=0,ignore_index=True,sort=False)

    if i == length: #if end reached
        df2 = pd.concat([df2,df[i-1]],axis=0,ignore_index=True,sort=False)   
```
